```python
import pandas as pd
from sklearn.datasets import load_iris

iris_load = load_iris()
iris = pd.DataFrame(iris_load.data, columns=iris_load.feature_names)
iris['Class'] = iris_load.target

# 숫자 레이블을 문자열 범주로 변환
iris['Class'] = iris['Class'].map({
    0: 'Setosa',
    1: 'Versicolor',
    2: 'Virginica'
})

# 여기서는 나중에 사용하기 위해 iris 사본을 저장해 둡니다.
iris_origin = iris.copy()
```

```python
class_group = {
    'Setosa': 'Group_A',
    'Versicolor': 'Group_B',
    'Virginica': 'Group_B'
}

iris['Class_group'] = iris['Class'].map(class_group)

iris[['Class', 'Class_group']].head()
```

```python
iris['Class_label2'] = pd.factorize(iris['Class'])[0]
iris[['Class', 'Class_label2']].head()
```

```python
# 예시: 품종 이름의 이니셜 기준 그룹화
iris['Class_initial'] = iris['Class'].str[0]
iris[['Class', 'Class_initial']].head()
```

```python
iris['petal_length_group'] = pd.cut(
    iris['petal length (cm)'],
    bins=[0, 2, 5, 10],
    labels=['Short', 'Medium', 'Long']
)

iris[['petal length (cm)', 'petal_length_group']].head()
```

```python
iris_dummy = pd.get_dummies(iris_origin, columns = ['Class'])
iris_dummy.head()
```

```python
iris_onehot = pd.get_dummies(
    iris_origin, columns=['Class'], drop_first=True
)
iris_onehot.head()
```
