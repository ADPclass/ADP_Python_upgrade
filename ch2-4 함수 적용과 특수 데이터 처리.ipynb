#### 2. 문자열 데이터 처리

```python
# 예제 데이터 선언
import pandas as pd
import numpy as np

score = {
    '성별': ['여자', '남자', '여자', '남자', '남자'],
    '성적': [70, 80, 90, 50, 100],
    '가산점': [5, 10, 0, 10, 5]
}
score = pd.DataFrame(score)
score
```

```python
score[['성적', '가산점']].apply(np.mean)
```

```python
score_pivot = score.pivot_table(columns='성별', values='성적')
score_pivot.apply(np.mean, axis=0)
```

```python
def final_score(row):
    total = row['성적'] + row['가산점']
    return min(total, 100)

score['최종점수'] = score.apply(final_score, axis=1)
score
```

```python
gender_map = {'남자': 'M', '여자': 'F'}
score['성별코드'] = score['성별'].map(gender_map)
score
```

```python
score['합격여부'] = score['최종점수'].map(lambda x: '합격' if x >= 80 else '불합격')
score
```

```python
import pandas as pd

landmark = pd.DataFrame({
    'name': ['경복궁', '불국사', '첨성대', '광한루원', '창경궁'],
    'english name': [
        'Gyeongbokgung Palace',
        'bulguksa',
        'cheomseongdae observatory',
        'gwanghanru',
        'Changgyeonggung Palace'
    ],
    'location': [
        '서울특별시 종로구 사직로 161',
        '경북 경주시 불국로 385',
        '경북 경주시 인왕동 839-1',
        '전북 남원시 요천로 1447',
        '서울 종로구 창경궁로 185'
    ]
})
```

```python
landmark['name_upper'] = landmark['english name'].str.upper()
landmark['name_lower'] = landmark['english name'].str.lower()
landmark
```

```python
# 하나의 컬럼에 공백제거를 수행하는 방법
landmark['english name'] = landmark['english name'].str.strip()

# 모든 컬럼에 공백 제거를 수행하는 방법
for col in landmark.columns:
    landmark[col] = landmark[col].str.strip()
```

```python
landmark['location'].str[3:6]
```

```python
landmark['location'].str.split().str.get(1)
```

```python
landmark['name_replace'] = landmark['english name'].str.replace(' ', '_', regex=False)
landmark['simple'] = (
    landmark['location']
    .str.replace('서울특별시', '서울', regex=False)
    .str.replace('경북 경주시', '경주', regex=False)
    .str.replace('전북 남원시', '남원', regex=False)
)
landmark[['english name', 'name_replace', 'location', 'simple']]
```

```python
landmark['location'].str.extract('(경주)')
```

```python
landmark['location'].str.extract(r'(\d+-?\d*)')
```

```python
seoul = landmark[landmark['location'].str.contains('서울', regex=False)]
seoul
```

```python
landmark['region'] = landmark['location'].str.extract(r'^(서울|경북|전북)')
landmark['is_seoul'] = landmark['location'].str.contains('서울', regex=False)
landmark[['location', 'region', 'is_seoul']]
```

```python
landmark['address_number'] = (
    landmark['location']
    .str.extract(r'(\d+)')
    .astype(int)
)
landmark[['address_number', 'location']]
```

```python
landmark['english name'].str.contains('Palace', regex=False)
```

```python
import numpy as np
import pandas as pd

landmark = pd.DataFrame({
    'name': ['경복궁', '불국사', '첨성대', '광한루원', None],
    'english name': [
        'Gyeongbokgung Palace',
        None,
        'cheomseongdae observatory',
        'gwanghanru',
        'Changgyeonggung Palace'
    ],
    'location': [
        '서울특별시 종로구 사직로 161',
        '경북 경주시 불국로 385',
        '경북 경주시 인왕동 839-1',
        None,
        '서울 종로구 창경궁로 185'
    ]
})

landmark.loc[3, 'name'] = '삭제'
landmark.loc[2, 'location'] = '삭제'

landmark
```

```python
landmark.isna()
```

```python
# 'name' 컬럼에서 '삭제' 값을 포함한 행 제거
landmark[landmark['name'] != '삭제']
```

```python
# 'english name' 컬럼에서 결측치(NaN) 행 제거
landmark.dropna(subset=['english name'])
```

```python
# 'english name' 컬럼에서 결측치를 'unknown'으로 대체
landmark['english name'] = landmark['english name'].fillna('삭제')

# "삭제"와 같은 문자열을 None으로 대체
landmark['name'] = landmark['name'].replace('삭제', None)

landmark
```

```python
landmark['english name'].str.contains('Palace', na=False)
```
