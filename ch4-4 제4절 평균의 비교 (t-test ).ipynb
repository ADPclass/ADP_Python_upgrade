```python
import pandas as pd
import scipy.stats as stats
# 1. 데이터 불러오기 및 A사 데이터 추출
df = pd.read_csv('https://raw.githubusercontent.com/ADPclass/ADP_book_ver01/main/data/delivery_data.csv')
a_time = df[df['company'] =='A']['delivery_time']
# 2. 정규성 검정 (Shapiro-Wilk)
# 귀무가설: "데이터가 정규분포를 따른다"
stat, p_norm = stats.shapiro(a_time)
print(f"정규성 검정 p-value: {p_norm:.4f}")
```

```python
# 기준값(mu)을 136으로 설정하여 검정 수행
mu_0 =136
t_stat, p_value = stats.ttest_1samp(a_time, popmean=mu_0)
print(f"검정통계량(t): {t_stat:.4f}")
print(f"p-value: {p_value:.4f}")
```

```python
import pandas as pd
import scipy.stats as stats
# 1. 데이터 불러오기 및 집단 분리
a_time = df[df['company'] == 'A']['delivery_time']
b_time = df[df['company'] == 'B']['delivery_time']

# 2. 정규성 검정 (Shapiro-Wilk)
print("A사 정규성 p-value:", stats.shapiro(a_time).pvalue)
print("B사 정규성 p-value:", stats.shapiro(b_time).pvalue)

# 3. 등분산성 검정 (Levene)
# 귀무가설: "두 집단의 분산은 같다."
stat, p_levene = stats.levene(a_time, b_time)
print(f"등분산성 검정 p-value: {p_levene:.4f}")
```

```python
# 정규성이 깨졌으므로 Mann-Whitney U 검정 실시
u_stat, p_val = stats.mannwhitneyu(a_time, b_time, alternative='two-sided')

print(f"검정통계량(U): {u_stat:.4f}")
print(f"p-value: {p_val:.4f}")
```

```python
import pandas as pd
import scipy.stats as stats
# 1. 데이터 생성 (Before, After)
data = {'before': [135, 140, 138, 142, 137, 139, 141, 136, 143, 138],
        'after': [128, 132, 129, 135, 130, 131, 133, 127, 136, 130]}
df_paired = pd.DataFrame(data)
# 2. 전후 차이 계산
diff = df_paired['after'] - df_paired['before']
# 3. 정규성 검정 (Shapiro-Wilk)
stat, p_norm = stats.shapiro(diff)
print(f"차이값의 정규성 p-value: {p_norm:.4f}")
```

```python
# 비모수 검정 (Wilcoxon Signed-Rank Test) 실시
# 두 집단 데이터를 직접 넣거나, 차이값(diff)을 넣어도 됩니다.
result = stats.wilcoxon(df_paired['after'], df_paired['before'], alternative='two-sided')

print(f"검정통계량: {result.statistic}")
print(f"p-value: {result.pvalue:.4f}")
```
