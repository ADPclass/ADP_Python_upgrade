```python
score = pd.DataFrame(
    {'국어': [100, 80], '수학': [75, 90], '영어': [90, 95]},
    index=['장화', '홍련']
)
score.loc['홍련', '영어'] = 100
score['국어'] = score['국어'] - 5
score
```

```python
new_students = pd.DataFrame(
    {'국어': [70, 85],'수학': [65, 100],'영어': [95, 65]},
    index=['콩쥐', '팥쥐']
)
score = pd.concat([score, new_students])

new_student = pd.Series(
    {'국어': 85, '수학': 55, '영어': 95},
    name='해님'
)
score = pd.concat([score, new_student.to_frame().T])
score
```

```python
new_student2 = {'국어': 75, '수학': 80, '영어': 80}
pd.concat([score, pd.DataFrame([new_student2])], ignore_index=True)
```

```python
science = [80, 70, 90, 85, 75]
score['과학'] = science
score['학년'] = 1
score
```

```python
score['과학'] = score['과학'] + 5
score['총점'] = score['국어'] + score['수학'] + score['영어'] + score['과학']
score
```

```python
score.drop('장화', inplace=True)
score.drop(columns=['과학', '학년', '총점'], inplace=True)
score
```

2. 조건식을 이용한 데이터 필터링



```python
import pandas as pd
students = pd.DataFrame({
    '이름': ['장화', '홍련', '콩쥐', '팥쥐', '해님', '달님'],
    '국어': [70, 85, None, 100, None, 85],
    '수학': [65, 100, 80, 95, None, 70]
})
students[students['이름'] == '장화']
```

```python
students[(students['국어'] >= 80) & (students['수학'] >= 80)]
```

```python
students.loc[6, '이름':'수학'] = ['별님', 50, 60]
students.loc[
    (students['국어'] >= 80) & (students['수학'] >= 70),
    '합격'
] = 'Pass'
students.loc[students['합격'] != 'Pass', '합격'] = 'Fail'
students
```

```python
import numpy as np
condition_list = [
    students['국어'] >= 90,
    (students['국어'] >= 80) & (students['국어'] < 90),
    (students['국어'] >= 70) & (students['국어'] < 80)
]
choice_list = ['A', 'B', 'C']
students['점수'] = np.select(condition_list, choice_list, default='F')
students
```

```python
students.isna().sum() # 열 기준
```

```python
students.isna().sum(axis=1)   # 행 기준
```

```python
students.dropna()          # 결측값이 포함된 모든 행 삭제
```

```python
students.dropna(thresh=4) # 결측값이 아닌 값이 4개 이상인 행만 유지
```
