# ch4-6 범주형 변수 간의 관계 분석 (카이제곱 검정)

## 적합성 검정

```python
import pandas as pd
# 데이터 불러오기
df = pd.read_csv("https://raw.githubusercontent.com/ADPclass/ADP_book_ver01/main/data/titanic.csv")
# titanic 데이터의 구조 확인
df.info()
```

```python
df_t = df[df['survived']==1]
table= df_t[['sex']].value_counts()
table
```

검정 결과, 카이제곱 통계량은 약 44.96이며 p-value는 유의수준 0.05보다 작게 나타나므로 귀무가
설을 기각합니다. 즉, 유의수준 0.05에서 타이타닉호 생존자의 남녀 비율은 50:50이라고 보기 어렵
습니다.

## 독립성 검정

```python
#유의수준 0.05로 적합도 검정 수행
from scipy.stats import chisquare
chi = chisquare(table, f_exp=[171,171])
print('<적합도 검정>￦n',chi)
```

```python
#데이터 불러오기
df = pd.read_csv("https://raw.githubusercontent.com/ADPclass/ADP_book_ver01/main/data/titanic.csv")
table = pd.crosstab(df['class'], df['survived'])
table
```

```python
# 카이제곱 검정을 통한 독립성 검정 수행
from scipy.stats import chi2_contingency
chi, p, df, expect = chi2_contingency(table) 
print('Statistic:', chi)
print('p-value:', p)
print('df:', df)
print('expect: ￦n', expect)
```

검정 결과 p-value가 유의수준 0.05보다 작게 나타나므로 귀무가설을 기각합니다. 따라서 유의수준 
0.05에서 좌석 등급과 생존 여부는 서로 독립이 아니며, 통계적으로 유의한 관련성이 존재한다고 할 
수 있습니다.
